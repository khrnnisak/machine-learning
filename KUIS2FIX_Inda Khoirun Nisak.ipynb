#### Preprocessing

##### Proses label encoding untuk data yang bersifat kategorial

```python
from sklearn import preprocessing
encode = preprocessing.LabelEncoder()

df['keterlibatan_vltr'] = encode.fit_transform(df['keterlibatan_vltr'])
df['skema'] = encode.fit_transform(df['skema'])
df['luaran_hki'] = encode.fit_transform(df['luaran_hki'])
df['jenis_luaran_hki'] = encode.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki'] = encode.fit_transform(df['potensi_hki'])
df['luaran_lain'] = encode.fit_transform(df['luaran_lain'])
df['jenis_luaran_lain'] = encode.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain'] = encode.fit_transform(df['status_luaran_lain'])
```

```python
df.head(100)
```

```python
# slicing data

X = df.drop(['keterlibatan_vltr'], axis= 1)
y = df['keterlibatan_vltr']
```

##### Data split (Training, Testing, Validation)

```python
# divide data into train, val, and test

from sklearn.model_selection import train_test_split

X_main, X_test, y_main, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_main, y_main, test_size=0.2, random_state=1, stratify=y_main)
```

#### Penerapan metode klasifikasi Decision Tree dan Support Vector Machine


##### Decision Tree

```python

from sklearn.tree import DecisionTreeClassifier # import DT
from sklearn.metrics import accuracy_score

df = DecisionTreeClassifier()

# Sesuaikan dt ke set training
df.fit(X_train, y_train)

# Memprediksi label set test
y_pred_df = df.predict(X_test)

#  menghitung set accuracy
acc = accuracy_score(y_test, y_pred_df)
print("Test set accuracy: {:.2f}".format(acc))
print(f"Test set accuracy: {acc}")
```

##### Support Vector Machine


```python
from sklearn.ensemble import RandomForestClassifier

rfm = RandomForestClassifier(n_estimators=10 , random_state=1)

# Sesuaikan dt ke set training
rfm.fit(X_train, y_train)

# Memprediksi label set test
y_pred_rfm = rfm.predict(X_test)

#  menghitung set accuracy
acc_rfm = accuracy_score(y_test, y_pred_rfm)
print("Test set accuracy: {:.2f}".format(acc_rfm))
print(f"Test set accuracy: {acc_rfm}")
```

#### Tunning Hyperparameter

##### Random Search DT

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import RepeatedKFold

rkf = RepeatedKFold(n_splits=4, n_repeats=3)
dt = DecisionTreeClassifier()
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11)) 
}
random = RandomizedSearchCV(dt, param_distributions=params, cv=rkf, n_jobs=-1)
%timeit random.fit(X, y)

# Evaluasi dengan score
score = random.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {random.best_params_}')
```

##### Grid Search

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold

rkf = RepeatedKFold(n_splits=4, n_repeats=3)
dt = DecisionTreeClassifier()
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11)) 
}
grid = GridSearchCV(dt, param_grid=params, cv=rkf, n_jobs=-1)
%timeit grid.fit(X, y)

# Evaluasi dengan score
gs = grid.score(X,y)

print(f'Hasil evaluasi: {gs}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```
