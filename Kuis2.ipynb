##### pengisian kolom yang bernilai Null dengan mean

```python
# fill null column id_ketua
mean_idket=df['id_ketua'].mean()
df['id_ketua'].fillna(value=mean_idket, inplace=True)

# fill null column id_anggota1
mean_id1=df['id_anggota1'].mean()
df['id_anggota1'].fillna(value=mean_id1, inplace=True)

# fill null column id_anggota2
mean_id2=df['id_anggota2'].mean()
df['id_anggota2'].fillna(value=mean_id2, inplace=True)

# fill null column id_anggota3
mean_id3=df['id_anggota3'].mean()
df['id_anggota3'].fillna(value=mean_id3, inplace=True)

# fill null column id_anggota4
mean_id4=df['id_anggota4'].mean()
df['id_anggota4'].fillna(value=mean_id4, inplace=True)
```

```python
df.isna().sum()
```

##### Dilakukan label encoder untuk data yang bersifat kategorial

```python
from sklearn import preprocessing
encode = preprocessing.LabelEncoder()

df['keterlibatan_vltr'] = encode.fit_transform(df['keterlibatan_vltr'])
df['skema'] = encode.fit_transform(df['skema'])
df['luaran_hki'] = encode.fit_transform(df['luaran_hki'])
df['jenis_luaran_hki'] = encode.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki'] = encode.fit_transform(df['potensi_hki'])
df['luaran_lain'] = encode.fit_transform(df['luaran_lain'])
df['jenis_luaran_lain'] = encode.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain'] = encode.fit_transform(df['status_luaran_lain'])
```

```python
df.head(100)
```

##### Proses slicing data

```python
# slicing data

X = df.iloc[:, 0:6].values
y = df['keterlibatan_vltr'].values
```

##### Menentukan data validasi

```python
# penentuan data validasi

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

# LR Object Definition
lr = LinearRegression()

# Fit model
lr.fit(X_train, y_train)

# Predict Test Data
y_pred = lr.predict(X_test)
```

##### Penentuan hasil akurasi menggunakan metode Decision Tree

```python

from sklearn.tree import DecisionTreeClassifier # import DT
from sklearn.metrics import accuracy_score

df = DecisionTreeClassifier()

# Sesuaikan dt ke set training
df.fit(X_train, y_train)

# Memprediksi label set test
y_pred_df = df.predict(X_test)

#  menghitung set accuracy
acc = accuracy_score(y_test, y_pred_df)
print("Test set accuracy: {:.2f}".format(acc))
print(f"Test set accuracy: {acc}")
```

##### Penentuan hasil akurasi menggunakan metode Support Vector Machine

```python
from sklearn.svm import SVC # import SVM classifier

svm_pol = SVC(kernel='poly')

# Fit ke model
svm_pol.fit(X_train, y_train)

# Prediksi
y_pred= svm_pol.predict(X_test)

# Evaluasi akurasi testing data
acc_svm = accuracy_score(y_test, y_pred)

# Print hasil evaluasi
print("Test set accuracy: {:.2f}".format(acc_svm))
print(f"Test set accuracy: {acc_svm}")
```

#### Tunning Hyperparameter dengan metode Decision Tree
##### Random Search

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import RepeatedKFold

rkf = RepeatedKFold(n_splits=4, n_repeats=3)
dt = DecisionTreeClassifier()
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11)) 
}
random = RandomizedSearchCV(dt, param_distributions=params, cv=rkf, n_jobs=-1)
%timeit random.fit(X, y)

# Evaluasi dengan score
score = random.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {random.best_params_}')
```

##### Grid Search

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold

rkf = RepeatedKFold(n_splits=4, n_repeats=3)
dt = DecisionTreeClassifier()
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11)) 
}
grid = GridSearchCV(dt, param_grid=params, cv=rkf, n_jobs=-1)
%timeit grid.fit(X, y)

# Evaluasi dengan score
gs = grid.score(X,y)

print(f'Hasil evaluasi: {gs}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```

#### Tunning Hyperparameter dengan metode Support Vector Machine 
##### Random Search

```python

from sklearn.svm import SVC

svm = SVC()

params = {
    'kernel': ['linear', 'poly', 'rbf'],
    'C': [1, 5, 10, 50, 100],
    'degree': [2, 3, 4, 5, 6, 7, 8, 9, 10],
}


random_sv = RandomizedSearchCV(svm, param_distributions=params, cv=rkf, n_jobs=-1)

%timeit random_sv.fit(X, y)

score = random_sv.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {random_sv.best_params_}')
```

##### Grid Search

```python

svm = SVC()

params = {
    'kernel': ['linear', 'poly', 'rbf'],
    'C': [1, 5, 10, 50, 100],
    'degree': [2, 3, 4, 5, 6, 7, 8, 9, 10],
}


grid_sv = RandomizedSearchCV(svm, param_grid=params, cv=rkf, n_jobs=-1)

%timeit grid_sv.fit(X, y)

score = grid_sv.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {grid_sv.best_params_}')
```
